```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
import math
median_bedrooms = math.floor(df.bedrooms.median())
median_bedrooms
```

```python
df.bedrooms = df.bedrooms.fillna(median_bedrooms)
df
```

```python
#training the model

reg = linear_model.LinearRegression()
reg.fit(df[['area', 'bedrooms', 'age']], df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3000,3,40]])
```

```python
# this is how it is calculated
112.06244194*3000+23388.88007794*3+-3231.71790863*40+221323.00186540396
```

```python
# this room is very new so the price of the home will be more because it has 5 years
reg.predict([[2500,4,5]])
```

```python
!pip install word2number

import pandas as pd
import numpy as np
from sklearn import linear_model
from word2number import w2n
```

```python
d = pd.read_csv('hiring.csv')
d
```

```python
d.experience = d.experience.fillna("zero")
d
```

```python
d.experience = d.experience.apply(w2n.word_to_num)
d
```

```python
import math
median_test_score = math.floor(d['test_score(out of 10)'].mean())
median_test_score
```

```python
d['test_score(out of 10)'] = d['test_score(out of 10)'].fillna(median_test_score)
d
```

```python
reg = linear_model.LinearRegression()
reg.fit(d[['experience','test_score(out of 10)','interview_score(out of 10)']],d['salary($)'])
```

```python
reg.predict([[2,9,6]])
```
